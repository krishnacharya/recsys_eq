```python
def plot_util(key:str, df): # for a fixed df as dimension varies
  utils_columns = ['avg_prod_util', 'min_prod_util', 'avg_user_util', 'min_user_util']
  
  dsp = ['dimension', 'seed', 'nprod']
  
  m = df[dsp + utils_columns].groupby(['dimension','nprod'], as_index=False).mean()
  s = df[dsp + utils_columns].groupby(['dimension','nprod'], as_index=False).std()
  nprods =  m['nprod'].unique()
  
  data_dim5 = {
    'x': nprods,
    'y': m[m['dimension'] == 5][key],
    'yerr': s[s['dimension'] == 5][key]}
  
  data_dim10 = {
    'x': nprods+0.3,   # adding offsets to x to make standard error bars non-overlapping
    'y': m[m['dimension'] == 10][key],
    'yerr': s[s['dimension'] == 10][key]}

  data_dim15 = {
    'x': nprods-0.3,
    'y': m[m['dimension'] == 15][key],
    'yerr': s[s['dimension'] == 15][key]}
  
  data_dim20 = {
    'x': nprods-0.3,
    'y': m[m['dimension'] == 20][key],
    'yerr': s[s['dimension'] == 20][key]}
  line_styles = [':o',':s' ,':o' , ':s'] # dotted circle and square

  
#   key_dim_mean = m[m['dimension'] == dim][key]
#   key_dim_std = s[s['dimension'] == dim][key]  
#   data = {'x': m['nprod'].unique(), 'y': key_dim_mean, 'yerr': key_dim_std}
  for idx, data in enumerate([data_dim5, data_dim10, data_dim15, data_dim20]):
    plt.errorbar(**data, fmt = line_styles[idx], capsize=3, capthick=1, elinewidth=1, \
                   alpha=0.9, markersize=4, label = f'dimension = {dims[idx]}')
  plt.legend(loc= "upper left")
  plt.xlabel("Number of producers")
  plt.ylabel("Average producer utility")
  plt.savefig(filename, bbox_inches='tight')
```
