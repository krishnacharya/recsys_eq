```python
import sys
sys.path.append('../source/')
import numpy as np
from Users import Users
from Producers import ProducersEngagementGame
from embeddings import *
from sklearn.preprocessing import normalize
from plotting import *
```

```python
# sorted user embeddings, linear probability, d 5
num_users = 10000
dimension = 5
num_producers = 100

user_embeddings = generate_uniform_users(dimension = dimension, num_users = num_users)
user_embeddings.sort(axis=1) # sort each row in ascending order

print(user_embeddings.shape)

PEng = ProducersEngagementGame(num_producers=num_producers, users = Users(user_embeddings), prob = 'linear')
NE, NE_compact, iters= PEng.best_response_dynamics(verbose=False) # runs a single best response dynamics

print(NE_compact, NE_compact.sum())

prod_dist = NE_compact / num_producers
user_dist = user_embeddings.sum(axis=0) / num_users
user_producer_2bar_plot(prod_dist, user_dist, '../plots/linear_sorted_dim5_100prod_10kusers.pdf')
```
